## <font color=#33A5FF>****2. Funciones básicas de lectura de ficheros****<tfont>

**index_col** = Cambiar el índice

```python
# Opción index_col
pd.read_excel("titanic.xlsx", index_col=0)
```

**nrows** = utilizar las n primeras filas

```python
# Opción nrows
pd.read_excel("titanic.xlsx",nrows=5)
```

**usecols**: Utilizar columnas específicas

```python
# Opción usecols
pd.read_excel("titanic.xlsx",usecols=["Survived","Nombre","Sexo"], index_col=0)
```

**skiprows**: saltarse filas de abajo hacia arriba

```python
# Opción skiprows
pd.read_excel("titanic.xlsx",skiprows=0)
```

**sheet_name**: Seleccionar las pestañas del libro excel

```python
# Opción sheet_name (pestañas o hojas de excel 0 o 1 o nombre de la hoja)
pd.read_excel("titanic.xlsx",sheet_name=0)
```

**header**: Se utiliza cuando las columnas no tienen encabezado

```python
# Opción header cuando no tienen encabezado
pd.read_excel("titanic.xlsx",header=0)
```

```python
# Opción heater asignando nombre de las columnas
pd.read_excel("titanic.xlsx",header=None, names=["Id_Pasajero","Survived","Clase","Nombre","Sexo","Edad","Familiares","Tiquete","Tarifa","Cabina","Embarcado"],sheet_name=1)
```

## <font color=#33A5FF>****3. Funciones básicas de almacenamiento****<tfont>

```python
# Ruta de la carpeta donde quiero guardar mis archivos exportados
titanic=pd.read_excel("titanic.xlsx")
```

```python
# Exportar archivo formato excel 
titanic.to_excel("titanic_exp.xlsx")
```

```python
# Exportar archivo formato excel sin índices
titanic.to_excel("titanic_exp.xlsx",index=False)
```

```python
# Exportar archivo formato .csv 
titanic.to_csv("titanic_exp.csv")
```

```python
# Exportar archivo formato .txt 
titanic.to_csv("titanic_exp.txt",sep=";")
```

```python
# Exportar archivo formato .html 
titanic.to_html("titanic_exp.html")
```
